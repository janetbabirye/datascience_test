```python
covid = pd.read_csv('COVID-19 Cases.csv')
covid_19 =covid.set_index('Date').head(10)
covid_19
```

# question 6

```python
covid_I = covid[(covid.Difference >0) & (covid.Case_Type == 'Confirmed') & (covid.Country_Region == 'Italy')]
covid_I.sort_values(by = ['Cases']).head()
```

```python
covid_G = covid[(covid.Difference >0) & (covid.Case_Type == 'Confirmed') & (covid.Country_Region == 'Germany')]
covid_G.sort_values(by = ['Cases']).head()
```
